```python
def compute_logistic(x,w,b):
    print("X:",x)
    print("w:",w)
    print("b:",b)
    print("Compute_linear :",compute_linear(x,w,b))
    print("Math :",math.e**(-compute_linear(x,w,b)))
    print("math+1:",1+(math.e**(-compute_linear(x,w,b))))
    print("Final :", 1/(1+(math.e**(-compute_linear(x,w,b)))))
    return 1/(1+(math.e**(-compute_linear(x,w,b))))
    
```

```python
compute_logistic(x,w,b)
```

```python
def compute_logistics_loss(x,y,w,b):
    row_count=x.shape[0]
    J_wb=0 
    J_wb+=(y*(np.log(compute_logistic(x,w,b)))+((1-y)*(np.log(1-compute_logistic(x,w,b)))))
    print("Row_count:",row_count)
    return np.sum(J_wb)/(-row_count)
```

```python
cost=compute_logistics_loss(x,y,w,b)
print("Cost :",cost)
#11.587969087943256
```

```python
def compute_dwj_jwb(w,y,b,j,x,m):
    return (np.sum(compute_logistic(x,w,b)-y*x[:,j]))/m
```

```python
def compute_db_jwb(w,y,b,x,m):
    return (np.sum(compute_logistic(x,w,b)-y))/m
```

```python
def compute_gradient(w,y,b,a,x):
    m=x.shape[0]
    b=b-(a*compute_db_jwb(w,y,b,x,m))
    for j in range(x.shape[1]):
        w[j]=w[j]-(a*compute_dwj_jwb(w,y,b,j,x,m))
    return w,b
```

```python
def compute_gradient_descent_2(w,y,b,a,x):
    
    previous_err= 0
    new_err= -1
    while abs(new_err-previous_err)>=0.00000001:
        print("previous_err",previous_err)
        print("new_err",new_err)
        previous_err=new_err
        w,b=compute_gradient(w,y,b,a,x)
        new_err=compute_logistics_loss(x,y,w,b)
        print("New :",new_err)
    return b,w
        
        
```

```python
x=np.random.rand(3,6)
w=initialize_weights(6)
b=random.randint(1,10)
y=np.array([0,1,0])
a=0.01
```

```python
def compute_differential_para(x,y,w,b):
    row_count,column_count=x.shape
    dj_db=0
    dj_dw=np.zeros((x.shape[1],))
    
    for row in range(row_count):
        err=(compute_logistic(x[row],w,b)-y[row])
        dj_db+=err
        for column in range(column_count):
            dj_dw[column]+=(err*x[row,column])
    dj_db/=row_count
    dj_dw/=row_count
    return dj_db,dj_dw
```

```python
dj_db,dj_dw=compute_differential_para(x,y,w,b)
print(dj_db,dj_dw)
#0.666666556330921 1.759265107992439
```

```python
def compute_gradient_descent(x,y,w,b,a):
    row_count,column_count=x.shape
    dj_db,dj_dw= compute_differential_para(x,y,w,b)
    alpha=a
    for count in range(1000000):
        w=w-(alpha*dj_dw)
        b=b-(alpha*dj_db)
    return b,w
```

```python
b,w=compute_gradient_descent(x_train,y_train,w,b,a)
print("b :",b)
print("w :",w)
cost=compute_logistics_loss(x_train,y_train,w,b)
print("Cost :",cost)
```
